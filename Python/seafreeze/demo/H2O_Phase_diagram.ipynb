Seafreeze can be used to create an H<sub>2</sub>O phase diagram in the region of the supported phases.  First, calculate the most stable phase over a grid of temperatures ranging from 200 - 300 K and for pressures up to 1 GPa.

```python
import os
if os.path.split(os.getcwd())[1] == 'demo':
    os.chdir(os.path.join(os.pardir, os.pardir))
```

```python
import numpy as np
import seafreeze.seafreeze as sf

P = np.arange(0, 2300, 0.5)
T = np.arange(200, 300, 0.1)
PT = np.array([P, T], dtype=object)
out = sf.whichphase(PT)
```

Plot each phase in a different color to form the phase diagram.  There are a lot of points to plot, so this may take a while.

```python
import matplotlib.pyplot as plt
%matplotlib inline

TT, PP = np.meshgrid(T, P)
fig, ax = plt.subplots()
for p in sf.phasenum2phaseDict.keys():
    if not np.isnan(p):
        pi = out==p
        ax.scatter(TT[pi], PP[pi], label=sf.phasenum2phase(p))
    
ax.legend(loc='upper right')
ax.set_xlabel('temperature (K)')
ax.set_ylabel('pressure (MPa)')
ax.set_title('$H_{2}O$ phase diagram')
```
